```python
# 必要なライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt

# x軸のデータを生成（0から10まで、50個の等間隔の数値）
x = np.linspace(0, 10, 50)

# y軸のデータを計算（線形関数）
y = 2 * x + 1

# グラフをプロット
plt.plot(x, y, '-r', label='y=2x+1')

# タイトルと軸ラベルの追加
plt.title('Graph of y=2x+1')
plt.xlabel('x')
plt.ylabel('y')

# 凡例の表示
plt.legend(loc='best')

# グリッドの表示
plt.grid(True)

# グラフの表示
plt.show()
```
